```python
import pytchat
import time
#Get PytchatCore object
#https://www.youtube.com/watch?v=pXmiczTCFg0
import glob
def get_link_youtube():
  url= ''
  while True:
    with open('/content/drive/MyDrive/DoAn/data/link_youtube.txt') as f:
      url = f.read()
    if url != '':
      break
  while True:
    try:
      with open('/content/drive/MyDrive/DoAn/data/link_youtube.txt') as f:
        url = f.read()
      livechat = pytchat.create(video_id = url)# video_id is https://....watch?v=Behind
      break
    except: 
      pass
  IN_PATH = '/content/drive/MyDrive/DoAn/data/youtube_cmt/'
  item_location=len(glob.glob(IN_PATH+'/*.json'))
  while livechat.is_alive():
      #Get chat data
      chatdata = livechat.get()
      
      for c in chatdata.items:
          print(f"{c.datetime} {c.author.name} {c.message} {c.amountString}")
          import json
  
          # Data to be written
          dictionary ={
              "name" : c.author.name,
              "time" : c.datetime,
              "message" : c.message,
              "amountString" : c.amountString
          }
            

            
          # Writing to sample.json
          with open("/content/drive/MyDrive/DoAn/data/youtube_cmt/file_{}.json".format(item_location), "w",encoding='utf-8') as outfile:
            json.dump(dictionary,outfile,ensure_ascii=False)

          item_location+=1

      time.sleep(3)
get_link_youtube()
```
